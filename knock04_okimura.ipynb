```python
#前処理
import MeCab

def create_MeCabFile(filename):
    with open(filename) as data_file, open(filename + '.mecab', 'w') as out_file:
        mecab = MeCab.Tagger()
        out_file.write(mecab.parse(data_file.read()))
```

```python
create_MeCabFile('neko.txt')
```

```python
#30
def read_MeCabFile(filename):
    with open(filename, encoding='utf-8') as mecabFile:
        sentense = []
        sentenses = []
        for morpheme in mecabFile.read().split('\n'):
            surfase = morpheme.split('\t')
            if len(surfase) >= 2:
                result = surfase[1].split(',')
                word = {
                    'surface' : surfase[0],
                    'base' : result[6],
                    'pos' : result[0],
                    'pos1' : result[1]
                }
                if word['pos1'] != '空白':
                    sentense.append(word)
                if word['pos1'] == '句点' or word['pos1'] == '空白':
                    sentenses.append(sentense)
                    sentense = []
        return sentenses
read_MeCabFile('neko.txt.mecab')
```

```python
result = read_MeCabFile('neko.txt.mecab')
```

```python
#31
def verb_MeCabFile(filename):
    list_verb = []
    for sentence in filename:
        for word in sentence:
            if word['pos'] == '動詞':
                list_verb.append(word['surface'])
    return list_verb
verb_MeCabFile(result)
```

```python
#32
def verb_base_MeCabFile(filename):
    list_verb = []
    for sentence in filename:
        for word in sentence:
            if word['pos'] == '動詞':
                list_verb.append(word['base'])
    return list_verb
verb_base_MeCabFile(result)
```

```python
#33
def sa_noun_MeCabFile(filename):
    list_noun = []
    for sentence in filename:
        for word in sentence:
            if word['pos1'] == 'サ変接続':
                list_noun.append(word['surface'])
    return list_noun
sa_noun_MeCabFile(result)
```

```python
#34
def no_noun_MeCabFile(filename):
    list_noun = []
    for sentence in filename:
        for i in range(len(sentence)- 1):
            if sentence[i]['pos'] == '名詞' and sentence[i + 1]['surface'] == 'の' and sentence[i +2]['pos'] == '名詞':
                list_noun.append(sentence[i]['surface'] + sentence[i + 1]['surface'] + sentence[i + 2]['surface'] )
    return list_noun
no_noun_MeCabFile(result)
```

```python
#35
def count_Mecabfile(filename, num = 100000):
    import collections
    lst_tmp = []
    for sentence in filename:
        for word in sentence:
            lst_tmp.append(word['surface'])
    cnt = collections.Counter(lst_tmp)
    return cnt.most_common(num)
count_Mecabfile(result)
```

```python
#36
def plot_fre(filename):
    from matplotlib import pyplot as plt
    from matplotlib.font_manager import FontProperties
    fp = FontProperties(fname='/System/Library/Fonts/ヒラギノ角ゴシック W2.ttc')
    dataset = list(zip(*count_Mecabfile(filename,10)))
    plt.xticks(range(len(dataset[0])), dataset[0], fontproperties=fp)
    plt.bar(range(len(dataset[0])), dataset[1], align='center')
plot_fre(result)
```

```python
#37
def co_occurance_neko(filename):
    import collections
    lst_occ = []
    for sentence in filename:
        if any(word['base'] == '猫' for word in sentence):
            words = [word['base'] for word in sentence if word['base'] != '猫']
            lst_occ.extend(words)
    occ = collections.Counter(lst_occ).most_common()
    return occ
co_occurance_neko(result)
```

```python
#38
def histogram(filename):
    from matplotlib import pyplot as plt
    from matplotlib.font_manager import FontProperties
    fp = FontProperties(fname='/System/Library/Fonts/ヒラギノ角ゴシック W2.ttc')
    dataset = list(zip(*count_Mecabfile(filename)))
    plt.xlabel('出現頻度', fontproperties=fp)
    plt.ylabel('単語の種類数', fontproperties=fp)
    plt.grid(axis='y')
    plt.xlim(xmin=1, xmax=20)
    plt.hist(dataset[1], bins=30, range=(1, 30))
histogram(result)
```
